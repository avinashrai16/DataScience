```python
class Node:

    def __init__(self, data=None, next=None):
        self._data =  data
        self._next = next
    
    # method to set the data value
    def setData(self, data):
        self._data = data
    
    # method to get the data value:
    def getData(self):
        return self._data
    
    # method to set the next:
    def setNext(self, next):
        self._next = next
    
    # method to get the next:
    def getNext(self):
        return self._next
    
    @property
    def data(self):
        return self._data
    
    @data.setter
    def data(self, upadated_data):
        self._data = upadated_data
    
    @property
    def next(self):
        return self._next
    
    @next.setter
    def next(self, upadated_next):
        self._next = upadated_next
    
    # create a linked list --> collection of list nodes
head = Node(1)
node2 = Node(2)
node3 = Node(3)
node4 = Node(4)
node5 = Node(5)
node6 = Node(6)
# creating the linkage
head.setNext(node2)
node2.setNext(node3)
node3.setNext(node4)
node4.setNext(node5)
node5.setNext(node6)
node6.setNext(node2)


def traverse_linked_list(head):
    temp = head
    while (temp):
        if(temp.getNext() == None): # for the tail of linked list, without end
            print(temp.data) # via property decorator
        else:
            print(temp.getData(), end="->")
        temp = temp.getNext() # go to the next node

def length_linked_list(head):
    temp = head
    len = 0
    while (temp):
        len += 1
        temp = temp.getNext() # go to the next node
    return len

#traverse_linked_list(head)
```

```python
# middle element of a linked list

def middle_element(head):
    pointer_1 = head # this pointer will jump 2 nodes at a time
    pointer_2 = head # this pointer will jump 1 node at a time

    while pointer_1 != None and pointer_1.getNext() != None:
        pointer_2 = pointer_2.getNext()
        pointer_1 = pointer_1.getNext().getNext()
        # print("pointer_1",pointer_1.getData())
        # print("pointer_2",pointer_2.getData())
    return pointer_2.getData()

middle_element(head=head)
```

```python
# detect the cycle in linke list
# this will run only for linked list without duplicate value
def isCyclePresent(head):
    slow = head
    fast = head

    while(fast and fast.getNext()):
        slow = slow.getNext()
        fast = fast.getNext().getNext()

        if fast and slow.getData() == fast.getData():
            return True #cycle exists
    return False #cycle does not exists

isCyclePresent(head)
```
