```python
# Example 1 
# Factorial of a given number

def fact(n): # T(n)
    if n <=1: # base case condtion(where the recussion has to be stopped) , time complexity is Constante i.e. C / O(1) Big O of 1
        print(n,end="=")
        return 1
    else:
        print(n,end="*")
        return n*fact(n-1) # recursive call, time complexity will be T(n-1)+C (C is for doing muliplication operarion)
try:
    input_num=int(input("enter the number for getting its factorial"))
    if input_num<0: # handle the negative value
        raise ValueError("negative value is not allowed")
    print("factorial of",input_num,"is:")
    print(fact(input_num))
except Exception as ex:
    print("Only non negative numbers are allowed, please inpur correct value.")

"""
The Recursion will be stored in Stack memory (LIFO " last in first out")

1.All the recursive code has Recurssion Realtion.
above code has following relations
T(n) = T(n-1)+C # T(n) is the time complexity function,
n*fact(n-1) # recursive call, time complexity will be T(n-1)+C (C is for doing muliplication operarion)
1.Time complexity: is T(n) = O(n) # BIG O of n

2.Space complexity: will be O(n) # BIG O of n , due to the fact we are calling fact(n) function n number of times , 
and the same will be occupied in Stack memory
Ex:if n=5 , then we are calling the fact function 5 times so space complexity will be based on n.

"""
```

```python
# Example 2
# Fibonacci Series

def fibonacci(n):
    if n<=1:
        return n # base case condtion(where the recussion has to be stopped) , time complexity is Constante i.e. C / O(1) Big O of 1
    else:
        return fibonacci(n-1)+fibonacci(n-2)

try:
    input_num=int(input("enter the number for getting its fibonacci"))
    if input_num<0: # handle the negative value
        raise ValueError("negative value is not allowed")
    print("fibonacci of",input_num,"is:")
    print(fibonacci(input_num))
except Exception as ex:
    print("Only non negative numbers are allowed, please inpur correct value.")

"""
The Recursion will be stored in Stack memory (LIFO " last in first out")

above code has following Recurssion relations
T(n) = T(n-1)+T(n-2)+C # T(n) is the time complexity function,
fibonacci(n-1)+fibonacci(n-2) # recursive call, time complexity will be T(n-1)+T(n-2)+C (C is for doing addtion operarion)

1.Time complexity: is T(n) = O(2^n) # BIG O of 2 to the power n , this is based on Recurrsive tress approach

2.Space complexity (how nuch space needed in stack memory): will be O(n) # BIG O of n , 
due to the fact we are calling fibonacci(n) function n number of times , 
and the same will be occupied in Stack memory
Ex:if n=5 , then we are calling the fibonacci function 5 times and it will have 5 stacks in stacks memory,
 so space complexity will be based on n.

"""
```

```python
# Example 3
# Climb Stairs with recurssion

def climbStairs(n):
    if(n<=3):
        return n
    return climbStairs(n-1)+climbStairs(n-2)

try:
    input_num=int(input("enter the number for climbing of stairs"))
    if input_num<0: # handle the negative value
        raise ValueError("negative value is not allowed")
    print("ways for climbing",input_num,"steps:")
    print(climbStairs(input_num))
except Exception as ex:
    print("Only non negative numbers are allowed, please inpur correct value.")


"""
Overlapping Subproblem : recomputaion of similar function causing TLE (Time Limit Exceeded), so for bigger n value we have have TLE error, 
hence recourssion is not efficient operation
Dynamic Programming will be used to overcome TLE but DP will increase Space compelxity

The Recursion will be stored in Stack memory (LIFO " last in first out")

above code has following Recurssion relations
T(n) = T(n-1)+T(n-2)+C # T(n) is the time complexity function,
climbStairs(n-1)+climbStairs(n-2) # recursive call, time complexity will be T(n-1)+T(n-2)+C (C is for doing addtion operarion)

1.Time complexity: is T(n) = O(2^n) # BIG O of 2 to the power n , this is based on Recurrsive trees approach

2.Space complexity (how nuch space needed in stack memory): will be O(n) # BIG O of n , 
due to the fact we are calling fibonacci(n) function n number of times , 
and the same will be occupied in Stack memory
Ex:if n=5 , then we are calling the fibonacci function 5 times and it will have 5 stacks in stacks memory,
 so space complexity will be based on n.

"""
```

```python
# Example 4
# Climb Stairs without recurssion

def climbStairs(n):
    if(n==1):
        return 1
    else:
        first = 1
        second = 2
        third =None
        for i in range (3,n+1):
            third = first+second
            first = second
            second = third
        return third
try:
    input_num=int(input("enter the number for climbing of stairs"))
    if input_num<0: # handle the negative value
        raise ValueError("negative value is not allowed")
    print("ways for climbing",input_num,"steps:")
    print(climbStairs(input_num))
except Exception as ex:
    print("Only non negative numbers are allowed, please inpur correct value.")

# Notes:
"""
1.Time complexity: is T(n) = O(n) # BIG O n , as there only one for loop
2.Space complexity (how nuch space needed in stack memory): will be constant as we are not adding any extra arrays , 
"""
```
