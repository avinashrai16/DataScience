```python
# example 1
input_num = 1234
def find_digit_sum(n):
    if n<10:
        return n
    return n%10+find_digit_sum(n//10)
print(find_digit_sum(input_num))

"""
Time complexity : 
T(n) = n%10 + find_digit_sum(n//10) # first part is constant hence for time complexity ignore first part and divsion on 10
T(n) = T(n-1)+c
T(n) = T(n-2)+2c
T(n) = T(n-3)+3c
so Time complexity will be Big O of n = O(n) # where n is number of digits


Spce complexity : it will be based on number of digits in the input number ,it will be O(n) # n is number of digits
"""
```

```python
# example2
# power of  a^n
def powerOf(a,n):
    if n == 0: # base condtion as a^0=1
        return 1
    else:
        return a*powerOf(a,n-1)
    
a= 2
n= 16
print(powerOf(a,n))

"""
Time complexity :
a*powerOf(a,n-1)
T(n) = T(n-1)+C # C is for multipilcation 
So Time complexity will be Big O of n O(n) as it will run same as per the value of Power

Spce complexity :it will be based on value of power ,it will be O(n) # n is power, we are running recurssion as per the number of Power
"""
```

```python
# example3
# power of  a^n , optimized code
def powerOf(a,n):
    if n == 1: # base condtion as a^1=a
        return a
    else:
        result = powerOf(a,n//2)
        finalResult = result*result
        if n%2==0:# if power is even we devide the power,we can double the result
            return finalResult
        else: # if power is odd we multiply the number with result of even power Ex. 2^17 = 2*2^16
            return a*finalResult
    
a= 2
n= 17
print(powerOf(a,n))

"""
Time complexity :
powerOf(a,n//2)
T(n) = T(n/2)+C # Equation1 C is for multiplication
T(n) = T(n/2^2)+2C # Equation2
T(n) = T(n/2^3)+3C # Equation3

K times 
T(n) = T(n/2^k)+kc 
 as per base case condtion n=1 retrun a
n = 2^k # take log on both side
k = log base of 2 of n
# put the valu of k to T(n) = T(n/2^k)+kc , and u will get T(n) as log of n base 2
So Time complexity will be Big O of log of n# O(log n) as it will run same as per the value of Power by 2

Spce complexity :it will be based on value of power ,it will be O(log n base 2) # n is power, 
we are running recurssion as per the number of Power and dividing on each step
EX: 2^16 : 2^16,2^8,2^4,2^2,2^1 approx number call is 4/5 which will be close to log base 2 of 16 = log base 2 2^4 = 4 log base2^2 = 4
"""
```
