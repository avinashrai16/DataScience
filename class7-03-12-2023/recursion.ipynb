```python
# reverse an array wuth signle array using 2 pointer approach 
def reverse_array(arr):
    # Initialize two pointers
    start = 0
    end = len(arr) - 1

    while start < end:
        # Swap elements at the start and end indices
        arr[start], arr[end] = arr[end], arr[start]

        # Move the pointers towards the center
        start += 1
        end -= 1

# Example usage:
my_array = [1, 2, 3, 4, 5]
print("Original Array:", my_array)

reverse_array(my_array)

print("Reversed Array:", my_array)
```

```python
# convert the above to a recursion
def reverse_array_recursive(arr, start, end):
    # base case
    if start <= end:
        arr[start], arr[end] = arr[end], arr[start] # swapping the values
        reverse_array_recursive(arr, start+1, end-1)

    else:
        return

my_array = [1, 2, 3, 4, 5]
print("Original Array:", my_array)

reverse_array_recursive(my_array, 0, len(my_array)-1)

print("Reversed Array:", my_array)
```

```python
# palindrome
def chk_plaindrome_list(lis1):
    start, end = 0, len(lis1) - 1

    while start < end:
        if lis1[start] != lis1[end]:
            return False

        start += 1
        end -= 1

    return True

my_list = [1, 2, 2, 1]
result = chk_plaindrome_list(my_list)
if result:
    print("The list is a palindrome.")
else:
    print("The list is not a palindrome.")
```

```python
def rotate_list_right(lst, k):
    n = len(lst)
    split_index = n - k
    rotated_list = lst[split_index:] + lst[:split_index]
    return rotated_list

my_list = [1, 2, 3, 4, 5]
rotated_list = rotate_list_right(my_list, 3)

print("Original List:", my_list)
print("Rotated List :", rotated_list)
```

```python
def rotate_list_left(lst, k):
    n = len(lst)
    k = k % n  # Ensure k is within the range of the list length

    # Split the list into two parts
    split_index = k

    # Rotate the list by swapping the two parts
    rotated_list = lst[split_index:] + lst[:split_index]

    return rotated_list

# Example usage:
my_list = [1, 2, 3, 4, 5]
rotated_list = rotate_list_left(my_list, 2)

print("Original List:", my_list)
print("Rotated List (2 places to the left):", rotated_list)
```

```python
# rotation of list using loops
def rotate_list_right_loop(lst, k):
    k = k % len(lst)

    for i in range (0, k):
       #Logic to rotate the array
        prev = lst[0]
        curr = lst[0]
        lst[0] = lst[len(lst)-1]
        j=1
        while(j<len(lst)):
            curr = lst[j]
            lst[j] = prev
            prev = curr
            j+=1

my_list = [1, 2, 3, 4]
rotate_list_right_loop(my_list, 2)
print("Original List:", my_list)
```
